## Machine Learning
### Fundamental concepts:

- We describe a general process.
    - we didn't explain ML in detail
- More simply, **making computers learn from data**
    - and observing them getting better and better
    - results: computers do things that they weren't explicitly told
- The field is vast (and expanding)
    - there are manu sub-fields, variations and algorithms  ... but the basis is still the same.

### Types of ML ALgorithms:
1. Supervised learning: 
    - we train the program on previously known (labelled) data
    - after training, we except it to make predictions on new data
    - ex: regression, classification
2. Unsupervised learning:
    - we leave the program to find patterns in data
    - ex: clustering analysis, dimensionality reduction
3. Reinforcement learning:
    - a form of supervised learning
    - the program learns continuously
    - ex: learning to play a game by observing other players, learning to drive a car
    
### Algorith by Task
1. Statistical algorithms:
2. Regression - predicting a continuous variable
3. Classification - predicting class labels
4. Clustering - finding compact groups of data points
5. Dimensionality reduction: simplifying the input data
6. Recommendation - suggest items for users
7. Optimization - minimize/ maximize a target function
8. Testing and improvement algorithms - helper algorithms to select, fine-tune and optimize other ML algorithms
9. ...and more
